### Task- To perform SQL on Vaccination dataset

```python
import pandas as pd
import sqlite3
```

```python
con= sqlite3.connect('data.sqlite')
data= pd.read_sql_query(""" SELECT * FROM sqlite_master""", con)
data.head()
```

```python
data= pd.read_sql_query(""" SELECT * FROM country_vaccinations """,con)
data.head(3)
```

#### Partially and Completely Vaccinated Population in each Country

```python
vaccines= pd.read_sql_query(""" SELECT country, SUM(total_vaccinations) AS Total, SUM(people_vaccinated) AS Number_of_people_partially_vaccinated, SUM(people_fully_vaccinated) AS Number_of_people_fully_vaccinated FROM country_vaccinations GROUP BY country  """,con)
vaccines.head(10)
```

#### Total Vaccines Available in each Country

```python
diff_vacs= pd.read_sql_query(""" SELECT country, vaccines, COUNT(vaccines) AS Total FROM country_vaccinations WHERE country IN ("India","Japan","China","Canada","Russia","United States","Israel","Australia","Saudi Arabia", "France","Italy","Sri Lanka") GROUP BY country """,con)
diff_vacs
```

#### Monthly Vaccination in India

```python
monthwise= pd.read_sql_query(""" SELECT country,month, SUM(total_vaccinations_per_hundred) AS Vaccination_per_hundred  FROM MONTH_Sheet1 WHERE country = "India" AND month NOT IN ("September","October","November","December") GROUP BY month """,conn)
monthwise
```

#### Top 10 countries with Highest Daily Vaccination

```python
daily= pd.read_sql_query("""SELECT country, SUM(daily_vaccinations_per_million) AS daily_vaccinations_per_million FROM country_vaccinations GROUP BY country ORDER BY SUM(daily_vaccinations_per_million) DESC LIMIT 10 """,con)
daily
```

#### Health Agency of each Country

```python
source=pd.read_sql_query(""" SELECT country, source_name FROM country_vaccinations GROUP BY source_name LIMIT 10 """,con)
source
```

#### Top 10 Vaccine Supplying Companies

```python
available= pd.read_sql_query(""" SELECT vaccines AS vaccine_availability, source_name FROM country_vaccinations GROUP BY vaccines ORDER BY SUM(daily_vaccinations_per_million) DESC LIMIT 10 """,con)
available
```

#### Top 10 Countries with Highest Vaccination per hundred 

```python
hundred= pd.read_sql_query(""" SELECT country, SUM(people_vaccinated_per_hundred) AS vaccination_per_hundred , SUM(people_fully_vaccinated_per_hundred) AS complete_vaccination_per_hundred FROM country_vaccinations GROUP BY country ORDER BY SUM(people_fully_vaccinated_per_hundred) DESC LIMIT 10 """,con)
hundred
```
